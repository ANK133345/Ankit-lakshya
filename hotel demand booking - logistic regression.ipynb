```python
df[['agent','company']] = df[['agent','company']].fillna(0.0)
df['children'].fillna(round(df.children.mean()), inplace=True)
df['country'].fillna(df.country.mode().to_string(), inplace=True)
```

```python
df.isnull().sum()
```

```python
inputs  = df.drop('is_canceled',axis=1)
target = df['is_canceled']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target, test_size=0.33, random_state=42)
```

```python
X_train_numeric = X_train.select_dtypes(include = np.number)
X_test_numeric = X_test.select_dtypes(include = np.number)
```

```python
X_train_numeric.shape
```

```python
X_test_numeric.shape
```

```python
X_train_cat = X_train.select_dtypes('object')
X_test_cat = X_test.select_dtypes('object')
```

```python
X_train_cat.shape
```

```python
X_test_cat.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train_numeric)
scaler.fit(X_test_numeric)
```

```python
X_train_scaled = scaler.transform(X_train_numeric)
X_test_scaled = scaler.transform(X_test_numeric)
```

```python
X_train_scaled.shape
```

```python
X_test_scaled.shape
```

```python
le = LabelEncoder()
```

```python
X_train_enc = X_train_cat.apply(LabelEncoder().fit_transform).to_numpy()
X_test_enc = X_test_cat.apply(LabelEncoder().fit_transform).to_numpy()
```

```python
X_train_transformed = np.concatenate([X_train_numeric,X_train_enc],axis=1)
X_test_transformed = np.concatenate([X_test_numeric,X_test_enc],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear')
```

```python
model.fit(X_train_transformed,y_train)
```

```python
model_pred = model.predict(X_test_transformed)
```

```python
model.score(X_train_transformed,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,model_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
lr_matrix=confusion_matrix(y_test,model_pred)
sns.heatmap(lr_matrix , cmap = 'gist_earth',annot = True)
```
