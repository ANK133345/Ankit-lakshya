```python
import seaborn as sns
sns.set_style('darkgrid')
import matplotlib.pyplot as plt
%matplotlib inline
for c in df.columns[1:]:
    df.boxplot(c,by = 'Class',figsize = (7,4),fontsize=10)
plt.xlabel('WineClass')
plt.title("{}\n".format(c),fontsize=13)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
X = scaler.fit_transform(X)
```

```python
dfx = pd.DataFrame(data=X,columns=df.columns[1:])
```

```python
dfx.head(10)
```

```python
dfx.describe()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
pca.fit(dfx)
x_pca = pca.transform(dfx)
x_pca.shape
```

```python
dfx_trans = pd.DataFrame(data= x_pca)
dfx_trans.head(10)
```

```python
plt.figure(figsize =(8, 6))
sns.scatterplot(dfx_trans[0],dfx_trans[1],c = df['Class'],cmap ='cividis')
plt.title("Class separation using first two principal components\n",fontsize=20)
plt.xlabel("Principal component-1",fontsize=15)
plt.ylabel("Principal component-2",fontsize=15)
plt.show()
```
