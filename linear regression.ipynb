```python
from sklearn.linear_model import LinearRegression
```

```python
model.fit(x,y)
```

```python
predictions
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(y,predictions)
```

```python
rmse =np.sqrt(mean_squared_error(y,predictions))
```

```python
rmse
```

```python
model.coef_
```

```python
model.intercept_
```

```python
r2_score(y,predictions)
```

```python
def estimate_charges(x, w, b):
    return x * w + b
```

```python
def try_parameters(w, b):
    x = non_smoker[['age']]
    y = non_smoker.charges
    predictions = estimate_charges(x, w, b)
    
    plt.plot(x, predictions, 'r', alpha=0.9);
    plt.scatter(x,y, s=8,alpha=0.8);
    plt.xlabel('Age');
    plt.ylabel('Charges')
    plt.legend(['Prediction', 'Actual']);
    
    loss = np.sqrt(mean_squared_error(y, predictions))
    print("RMSE Loss: ", loss)
```

```python
try_parameters(model.coef_,model.intercept_)
```

```python
# alternative method
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_new = non_smoker
```

```python
df_new.head(2)
```

```python
df_new.tail(2)
```

```python
df_new.charges
```

```python
sns.boxplot(df_new.charges)
```

```python
sns.boxplot(df_new.age)
```

```python
percentile25 = df_new['charges'].quantile(0.25)
percentile75 = df_new['charges'].quantile(0.75)
```

```python
iqr = percentile75-percentile25
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
df_new[df_new['charges'] > upper_limit]
```

```python
df_new[df_new['charges'] < lower_limit]
```

```python
df2 = df_new[df_new['charges'] < upper_limit]
```

```python
df2.head()
```

```python
sns.scatterplot(x = df2.age ,y = df2.charges)
```

```python
X = df2[['age']]
Y = df2.charges
```

```python
regression_model = LinearRegression()
```

```python
regression_model.fit(X,Y)
```

```python
y_predicted = regression_model.predict(X)
```

```python
y_predicted
```

```python
regression_model.coef_
```

```python
regression_model.intercept_
```

```python
mse = mean_squared_error(Y, y_predicted)
```

```python
mse
```

```python
RMSE =np.sqrt(mean_squared_error(Y, y_predicted))
print(RMSE)
```

```python
r2_score(Y,y_predicted)
```

```python
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error: ', RMSE)
print('r2 score: ', r2_score)
```
