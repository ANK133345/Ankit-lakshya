```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\excel\weatherAUS.csv')
```

```python
df
```

```python
df.drop(['Date'],axis = 1,inplace = True)
```

```python
categorical_df = df.select_dtypes(include = ['object']).copy()
numerical_df = df.select_dtypes(exclude = ['object']).copy()
```

```python
categorical_df
```

```python
print(categorical_df['WindGustDir'].value_counts())
```

```python
numerical_df
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
imputer.fit(numerical_df)
```

```python
numerical_transform = imputer.transform(numerical_df)
```

```python
numerical_transform
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(numerical_transform)
```

```python
numerical_scaled = scaler.transform(numerical_transform)
```

```python
numeric = pd.DataFrame(numerical_scaled,columns= [numerical_df.columns])
```

```python
numeric
```

```python
categorical_encoded = pd.get_dummies(categorical_df,drop_first = True)
```

```python
categorical_encoded
```

```python
categorical_encoded.isna().sum()
```

```python
df_transformed = pd.concat([numeric,categorical_encoded],axis=1)
```

```python
df_transformed
```

```python
inputs = df_transformed.iloc[:,:-1]
targets = df_transformed.iloc[:,-1]
```

```python
inputs
```

```python
targets
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,targets, test_size=0.33, random_state=42)
```

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier_rf = RandomForestClassifier(max_depth = 4)
```

```python
classifier_rf.fit(X_train, y_train)
```

```python
classifier_rf.score(X_train,y_train)
```

```python
pred_rf=classifier_rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred_rf)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize = (15,10))
tree.plot_tree(classifier_rf.estimators_[0],filled = True,feature_names = X_train.columns)
```
