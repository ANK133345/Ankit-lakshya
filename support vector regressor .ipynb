```python
inputs = df.iloc[:,0:8]
targets =df['median_house_value']
```

```python
inputs
```

```python
targets
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,targets, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train,y_train)
```

```python
s1 = svr.score(X_train, y_train)
s2 = svr.score(X_test, y_test)
```

```python
# model performa horriblly
print("RÂ² of Support Vector Regressor on training set: {:.3f}".format(s1))
print("RÂ² of Support Vector Regressor on test set: {:.3f}".format(s2))
```

```python
Y_train = y_train.values.reshape(-1,1)
Y_test = y_test.values.reshape(-1,1)
```

```python
#lets do feature scaling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
y_train_scaled = scaler.fit_transform(Y_train)
y_test_scaled = scaler.fit_transform(Y_test)
```

```python
svr1 = SVR()
svr1.fit(X_train_scaled,y_train_scaled)
```

```python
s3 = svr1.score(X_train_scaled, y_train_scaled)
s4 = svr1.score(X_test_scaled, y_test_scaled)
```

```python
# underfiiting problem performance is good on training and  bad on test data
print("RÂ² of Support Vector Regressor on training set: {:.3f}".format(s3))
print("RÂ² of Support Vector Regressor on test set: {:.3f}".format(s4))
```

```python
svr2 = SVR(C=2.0)
svr2.fit(X_train_scaled, y_train_scaled)
s5 = svr2.score(X_train_scaled, y_train_scaled)
s6 = svr2.score(X_test_scaled, y_test_scaled)
print("RÂ² of Support Vector Regressor on training set: {:.3f}".format(s5))
print("RÂ² of Support Vector Regressor on test set: {:.3f}".format(s6))
```
