```python
x = df.iloc[:,3:7]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm
```

```python
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
coef = pd.DataFrame(data = lm.coef_ , index = X_train.columns ,columns = ['coefficients'])
```

```python
coef
```

```python
y_prediction = lm.predict(X_test)
```

```python
y_prediction
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_prediction)
RMSE = np.sqrt(mean_squared_error(y_test, y_prediction))
```

```python
mse
```

```python
RMSE
```

```python
r2_score(y_test,y_prediction)
```

```python
#plot of residuals
plt.figure(figsize = (7,5))
sns.distplot(y_test-y_prediction)
plt.xlabel("Residuals",fontsize=12)
plt.title("Histogram of residuals to check for normality",fontsize=12)
```

```python
plt.figure(figsize = (7,5))
sns.scatterplot(y_test,y_prediction)
plt.xlabel("Actual Amount",fontsize=12)
plt.ylabel("Predicted Amount", fontsize=12)
plt.title('Actual vs predicted')
```

```python
plt.figure(figsize=(7,5))
residuals = y_test-y_prediction
sns.scatterplot(y_prediction,residuals)
plt.title("Residuals vs. predicted values plot (Homoscedasticity)\n",fontsize=12)
plt.xlabel("Prediction",fontsize=12)
plt.ylabel("Residuals", fontsize=12)
```
